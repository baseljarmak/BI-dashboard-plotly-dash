```python
import math
import plotly.graph_objects as go
import pandas as pd
import numpy as np
#below code taken from https://github.com/serge-tochilov/barchart3d-plotly/blob/master/barchart3d.py on 280824
import matplotlib
import plotly.express as px
import itertools

def barchart3d(labels, z_data, title, z_title,
               n_row=0, width=900, height=900, thikness=0.3, colorscale='blues', color='blue',
               **kwargs):
    """
    Draws a 3D barchart
    :param labels: Array_like of bar labels
    :param z_data: Array_like of bar heights (data coords)
    :param title: Chart title
    :param z_title: Z-axis title
    :param n_row: Number of x-rows
    :param width: Chart width (px)
    :param height: Chart height (px)
    :param thikness: Bar thikness (0; 1)
    :param colorscale: Barchart colorscale
    :param **kwargs: Passed to Mesh3d()
    :return: 3D barchart figure
    """

    to_add_trace = []
    if n_row < 1:
        n_row = math.ceil(math.sqrt(len(z_data)))
    thikness *= 0.3

    for iz, z_max in enumerate(z_data):
        x_cnt, y_cnt = iz % n_row, iz // n_row
        x_min, y_min = x_cnt - thikness, y_cnt - thikness/6
        x_max, y_max = x_cnt + thikness, y_cnt + thikness
        print("x_cnt={}, y_cnt={}".format(x_cnt, y_cnt))
        print("x_min={}, y_min={}".format(x_min, y_min))
        print("x_max={}, y_max={}".format(x_max, y_max))

        to_add_trace.append(go.Mesh3d(
            x=[x_min, x_min, x_max, x_max, x_min, x_min, x_max, x_max],
            y=[y_min, y_max, y_max, y_min, y_min, y_max, y_max, y_min],
            z=[0, 0, 0, 0, z_max, z_max, z_max, z_max],
            alphahull=0,
            #intensity=[0, 0, 0, 0, z_max, z_max, z_max, z_max], affects all bars having d same color
            coloraxis='coloraxis',
            hoverinfo='skip',
            **kwargs))

        
   
    # mesh3d doesn't currently support showLegend param, so
    # add invisible scatter3d with names to show legend
    for i, label in enumerate(labels):
        to_add_trace.append(go.Scatter3d(
            x=[None], y=[None], z=[None],
            #opacity=0, affects the label
            opacity=1,
            name=f'#{i+1} {label}'))

    return to_add_trace

def barchart3d_part(z_data, title, z_title,
               n_row=0, width=900, height=900, thikness=0.3, colorscale='blues', color='blue',
               **kwargs):

    to_add_trace = []
    if n_row < 1:
        n_row = math.ceil(math.sqrt(len(z_data)))
    thikness *= 0.3

    for iz, z_max in enumerate(z_data):
        x_cnt, y_cnt = iz % n_row, iz // n_row
        x_min, y_min = x_cnt - thikness, y_cnt - thikness
        x_max, y_max = x_cnt + thikness, y_cnt + thikness

        to_add_trace.append(go.Mesh3d(
            z=[0, 0, 0, 0, z_max, z_max, z_max, z_max],
            #intensity=[0, 0, 0, 0, z_max, z_max, z_max, z_max],
            **kwargs))
   
    return to_add_trace

def frame_args(duration):
    return {
        "frame": {"duration": duration},
        "mode": "immediate",
        "fromcurrent": True,
        "transition": {"duration": duration, "easing": "linear"},
    }

def get_annotation(zdata, labels):
    ann = []
    n_row = 0
    n_row = math.ceil(math.sqrt(len(zdata.to_list())))
    #n_row = 5
    thikness=0.7
    #labels=sliced_df_2016['fctr_attr'].unique().tolist()
    labels=labels
    width=900
    height=900
    z_title=""
    title=""
    for iz, z_max in enumerate(zdata.to_list()):
        x_cnt, y_cnt = iz % n_row, iz // n_row
        x_min, y_min = x_cnt - thikness, y_cnt - thikness
        x_max, y_max = x_cnt + thikness, y_cnt + thikness
        ann.append(dict(
                showarrow=False,
                x=x_cnt, y=y_cnt, z=z_max,
                text=f'<b>#{iz+1}</b>',
                font=dict(color='white', size=11),
                bgcolor='rgba(0, 0, 0, 0.3)',
                xanchor='center', yanchor='middle',
                hovertext=f'{z_max} {labels[iz]}'))
    return ann

def get_annotation_part(zdata, labels):
    ann = []
    n_row = 0
    n_row = math.ceil(math.sqrt(len(zdata.to_list())))
    #n_row = 5
    thikness=0.7
    #labels=sliced_df_2016['fctr_attr'].unique().tolist()
    labels=labels
    width=900
    height=900
    z_title=""
    title=""
    for iz, z_max in enumerate(zdata.to_list()):
        x_cnt, y_cnt = iz % n_row, iz // n_row
        x_min, y_min = x_cnt - thikness, y_cnt - thikness
        x_max, y_max = x_cnt + thikness, y_cnt + thikness
        ann.append(dict(
                showarrow=False,
                x=x_cnt, 
                y=y_cnt, 
                z=z_max,
                text=f'<b>#{iz+1}</b>',
                bgcolor='rgba(0, 0, 0, 0.3)',
                hovertext=f'{z_max} {labels[iz]}'))
    return ann

df = pd.read_csv("app/data/Y_001011002001.csv", skiprows=2)

df = df[df.analysis_var == 'O/S Total']
df = df[df._TYPE_ == 111]
df.drop(['rprt_id', 'cube_ndx' , 'cube_ndx_name', 'Cube_Sub_ndx', 'Cube_Sub_ndx_name', 'fctr', 'fctr_lbl', 'fctr_fmt', 'fctr_seq', 'analysis_var', '_TYPE_'], axis=1, inplace=True)
#df = df[df.dpd_attr == 'Current'] #added since we are interested about the fctr_attr

data_period_names = []
for date_period in df['date'].unique():
    data_period_name = {}
    data_period_name['name'] = eval(f"f'sliced_df_{date_period}'")
    data_period_name['data'] = df[df['date'] == date_period]
    data_period_names.append(data_period_name)

#sliced_df['year_only'] = (temp:=pd.to_datetime(sliced_df['date'],format="%Y"))

#list_of_fctr = list(set(sliced_df_2016['fctr_attr'].tolist() + sliced_df_2017['fctr_attr'].tolist()+ sliced_df_2018['fctr_attr'].tolist() + sliced_df_2019['fctr_attr'].tolist()))
list_of_fctr = [data['data']['fctr_attr'] for data in data_period_names]
list_of_fctr = list(set(list(itertools.chain(*list_of_fctr))))
list_of_fctr.sort()

for item in list_of_fctr:
    new_row = { "fctr_attr": "any" }
    for data in data_period_names:
        if not data['data']['fctr_attr'].str.contains(item).any(): 
            new_row["fctr_attr"] = item 
            #sliced_df_2016.loc[len(sliced_df_2016)] = new_row
            #sliced_df_2016.append(new_row, ignore_index=True)
            data["data"] = pd.concat([data["data"], pd.DataFrame([new_row])], ignore_index=True)

fig = go.Figure()
#print(sliced_df_2016['fctr_attr'].unique().tolist())
#print(gf_2016_sliced['Count'].sum())

#SETTING INITIAL FRAME
#find_out_max_figure_frame
max_data=len(data_period_names)-1
for trace in barchart3d( 
                             list_of_fctr,
                             data_period_names[max_data]['data'].groupby('fctr_attr')['Count'].sum(), #commented out bjarmak 041220                                                                                   
                             "", "", opacity=0.6, flatshading=True, n_row=0
         ):
    fig.add_trace(trace)


#SETTING FRAMES
for i, data in enumerate(data_period_names):
    data['data'].sort_values('fctr_attr', inplace=True)
    data['group'] = data['data'].groupby('fctr_attr')
    '''
    data['trace'] = go.Mesh3d(z=[data['group']['Count'].sum()],
                             alphahull=0,
                            intensity=[0, 0, 0, 0, z_max, z_max, z_max, z_max],
                            coloraxis='coloraxis',
                            hoverinfo='skip')
    '''
    data['trace'] = barchart3d_part(                                                                                                                             
                         data['group']['Count'].sum(), #commented out bjarmak 041220                                                                                   
                         "", "", n_row=0
     )
    data['annotations'] = get_annotation(data['group']['Count'].sum(), list_of_fctr)
    #print(data['trace'])


ann = get_annotation(data_period_names[0]['group']['Count'].sum(), list_of_fctr)
#print(ann)
#for i in range(0,3):
    #print(data_period_names[i]['annotations'])

fig.update(frames=[
    go.Frame(data=(data_period_names[i]['trace'])) for i in range(0,len(data_period_names))
])
#fig.update(frames=[
#    go.Frame(name=(data_period_names[i]['name'])) for i in range(1,len(data_period_names))
#])
fig.update(frames=[
    go.Frame(name="f_"+str(i)) for i in df['date'].unique()
])
#fig.update(frames=[
    #go.Frame(traces=str(i)) for i in range(0, len(df['date'].unique()))
 #   go.Frame(traces=[max_data])
#])
#fig.update(frames=[
#    go.Frame(baseframe=data_period_names[len(data_period_names)-1]['name'])
#])
#fig.update(frames=[
#    go.Frame(layout=go.Layout(annotations=(data_period_names[i]['annotations'] for i in range(0,len(data_period_names)))))
#])

for k in range(len(fig.frames)):
    fig.frames[k]['layout'].update(scene=dict(annotations=(data_period_names[k]['annotations']))) #, selector=1 , selector={"mode":"text"} , selector={"type":"mesh3d"}

sliders = [
    {
        "transition": {"duration": 100},
        "currentvalue": {"font":{"size":12}, "visible": True, "xanchor":"center"},
        "len": 0.9,
        "x": 0.1,
        "y": 0,
        "steps": [
            {
                "args": [[f.name], {
                            'mode': 'immediate',
                           'frame': { 'duration': 600, 'redraw': True },
                            'transition': {'duration': 200}
                        }
                ],
                "label": str(f.name),
                "method": "animate",
            }
            for k, f in enumerate(fig.frames)
        ],
    }
]

#for k in fig.frames:
    #print(k)

print(sliders)

n_row = 0
#n_row = 5
thikness=0.5
#labels=sliced_df_2016['fctr_attr'].unique().tolist()
width=900
height=900
z_title=""
title=""
fig.update_layout(
        width=width, height=height,
        title=title, title_x=0.5,
        scene=dict(
            xaxis=dict(title=''),
            yaxis=dict(title=''),
            zaxis=dict(title=""),
            annotations=ann
        ),
        coloraxis=dict(
            colorscale='blues_r',
            colorbar=dict(
                title=dict(
                    text=z_title,
                    side='right'),
                xanchor='right', x=1.0,
                xpad=0,
                ticks='inside')),
        legend=dict(
            yanchor='top', y=1.0,
            xanchor='left', x=0.0,
            bgcolor='rgba(0, 0, 0, 0)',
            itemclick='toggle',
            itemdoubleclick='toggleothers'),
        showlegend=True,
        updatemenus = [
        {
            "buttons": [
                {
                    "args": [None, {"frame": {"duration": 500, "redraw": True},
                                    "fromcurrent": True, "transition": {"duration": 300}, 'mode': 'immediate'}],
                    "label": "Play",
                    "method": "animate"
                },
                {
                    "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                      "mode": "immediate",
                                      "transition": {"duration": 0}}],
                    "label": "Pause",
                    "method": "animate"
                }
            ],
            "direction": "left",
            "pad": {"r": 10, "t": 87},
            "showactive": False,
            "type": "buttons",
            "x": 0.1,
            "xanchor": "right",
            "y": 0,
            "yanchor": "top"
        }
        ],
        sliders=sliders,
)

fig.show()

print(data_period_names[0]['group']['Count'].sum())
print(data_period_names[3]['group']['Count'].sum())
```
